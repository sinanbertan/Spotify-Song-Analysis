```python
#importing libraries:
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import ydata_profiling as ydp
```

```python
#loading the data:
df_tracks = pd.read_csv('tracks.csv')
df_tracks.tail()
```

```python
#exploring the data:
df_tracks.info()
```

```python
df_tracks.describe().T
```

```python
df_tracks.columns
```

```python
df_tracks.isnull().sum()
```

```python
#10 less popular songs:
sorted_tracks=df_tracks.sort_values(by="popularity", ascending=True).head(10)
sorted_tracks
```

```python
#10 most popular songs:
sorted_tracks=df_tracks.sort_values(by="popularity", ascending=False).head(10)
sorted_tracks
```

```python
#changing the index of the dataset:

df_tracks.set_index("release_date", inplace=True)
df_tracks.index=pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
#changing duration_ms to minutes:
df_tracks["duration_ms"]=df_tracks["duration_ms"]/60000

df_tracks.rename(columns={"duration_ms":"duration_min"}, inplace=True)
df_tracks.head()
```

```python
#pearson correlation:
corr=df_tracks.drop(["key", "mode", "explicit"], axis=1).corr(method="pearson")
corr
```

```python
#visualizing the correlation:
plt.figure(figsize=(15,10))
hm=sns.heatmap(corr, annot=True, cmap="coolwarm",fmt=".2f", linewidths=2,center=0)
hm.set_title("Correlation Heatmap between variables", fontsize=20)
hm.set_xticklabels(hm.get_xticklabels(), rotation=90) #rotating the x labels
```

```python
#a sample of the dataset:
sample_df=df_tracks.sample(int(0.02*len(df_tracks)))
```

```python
#correlation between loudness and energy:

plt.figure(figsize=(10,5))
sns.regplot(y="loudness", x="energy",color="c", marker="+" ,data=sample_df).set(title="Correlation between loudness and energy")
```

```python

plt.figure(figsize=(10,5))
sns.regplot(x="popularity", y="danceability",color="m", marker="+" ,data=sample_df).set(title="Correlation between danceability and popularity")
```

```python
#creating new column named "years":
df_tracks["dates"]=df_tracks.index.get_level_values("release_date")
df_tracks.dates=pd.to_datetime(df_tracks.dates)
years=df_tracks.dates.dt.year
df_tracks.head()
```

```python
#number of songs per year:
rs=sns.displot(years, bins=100,discrete=True,color="g", kde=True,kind="hist",aspect=2).set(title="Number of songs per year")
plt.xlabel("Year")
plt.ylabel("Number of songs")
```

```python
#Duration of songs per year:
total_duration=df_tracks.duration_min
total_duration.head()
```

```python

sns.set_style("darkgrid")
sns.lineplot(x=years,y=total_duration, color="m").set(title="Duration of songs per year")
plt.xticks(rotation=60)
```

```python
#importing second dataset:
df_features=pd.read_csv("SpotifyFeatures.csv")
df_features.tail()
```

```python
#exploring the data:
df_features.info()
```

```python
df_features.describe().T
```

```python
df_features.isnull().sum()
```

```python
#changing duration_ms to minutes:
df_features["duration_ms"]=df_features["duration_ms"]/60000
df_features.rename(columns={"duration_ms":"duration_min"}, inplace=True)
df_features.head()
```

```python
#duration of songs in different genres:
plt.figure(figsize=(10,5))
sns.barplot(y="genre", x="duration_min", data=df_features, palette="magma").set(title="Duration of songs in different genres")
plt.xticks(rotation=90)
plt.xlabel("Genres")
plt.ylabel("Duration of songs")
plt.title("Duration of songs in different genres")
```

```python
#correration between popularity and liveness:
sample_df=df_features.sample(int(0.02*len(df_features)))
plt.figure(figsize=(10,5))
sns.regplot(x="popularity",y="liveness",color="c",marker="p",data=sample_df).set(title="Correlation between popularity and liveness")
sns.set_style("darkgrid")
plt.xlabel("Popularity")
plt.ylabel("Liveness")
plt.title("Correlation between popularity and liveness")
```

```python
# What are most popular genres?
plt.figure(figsize=(8,8))
sns.barplot(y="genre", x="popularity", data=df_features, palette="magma").set(title="Most popular genres")
```

```python
#outliers in popularity:
plt.figure(figsize=(10,5))
sns.boxplot(x="popularity",y="genre", data=df_features, palette="magma").set(title="Outliers in popularity")
```
